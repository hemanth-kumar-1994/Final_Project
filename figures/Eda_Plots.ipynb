```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

LOADING DATA


```python
file_path = "../data/processed/master_dataset.csv"
    
df = pd.read_csv(file_path)
    
print("Data Loaded Successfully")
```

GLOBAL EV ADOPTION TREND

```python
""" 
This plot answers the question has EV adoption increased globally over time

"""

global_trend = df.groupby("year")["ev_sales_share"].mean()


plt.figure(figsize=(10,5))
plt.plot(global_trend.index, global_trend.values, marker="o")
plt.title("Average EV Sales Share Across Countries")
plt.xlabel("Year")
plt.ylabel("EV Sales Share (%)")
plt.grid(True)
plt.show()
```

TOP 20 COUNTRIES BY EV ADOPTION

```python
""" 
This plot shows which countries are the leaders in ev adoption

"""

latest_year = df["year"].max()

latest = df[df["year"] == latest_year]

top20 = (
    latest
    .sort_values(
        "ev_sales_share",
        ascending=False
    )
    .head(20)
)

plt.figure(figsize=(10,5))
plt.bar(top20["country"], top20["ev_sales_share"])
plt.xticks(rotation=60)
plt.title(f'Top 20 Countries By EV Sales Share  {latest_year}')
plt.ylabel("EV Sales Share (%)")
plt.show()
```
