```python
try:
    file_path = "../data/raw/EV Data Explorer 2026.xlsx"
    
    df = pd.read_excel(file_path,sheet_name="GEVO_EV_2026")
    
    print("Data Loaded Successfully")

except:
    print("File not found")
```

# ==========================================
# 3. UNDERSTAND DATA CONTENT
# ==========================================

```python
print("Number of Countries:")

print(df["region_country"].nunique())
```

# ==========================================
# 4. INVESTIGATE MISSING VALUES
# ==========================================

```python
print("Rows with Missing Country:")
print(
    df[df["region_country"].isna()] 
)
```

# ==========================================
# 5. BASIC PREPROCESSING
# ==========================================

```python
# Removing duplicates

df = df.drop_duplicates()
```

```python
# Standardizing column names

df.columns = (

    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

```python
#Removing rows without country

df = df[df["region_country"].notna()]
```

```python
print("Cleaned Shape:")

print(df.shape)
```

# ==========================================
# 6. SAVE CLEAN DATA
# ==========================================

```python
df.to_csv(
    "../data/processed/ev_clean_dataset.csv",
    index=False
)

print("Clean dataset saved successfully.")
```
