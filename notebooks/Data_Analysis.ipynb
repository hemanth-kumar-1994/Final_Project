```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

CREATING DATAFRAME


```python
try:
    file_path = "../data/raw/EV Data Explorer 2025.xlsx"
    
    df = pd.read_excel(file_path,sheet_name="GEVO_EV_2025")
    
    print("Data Loaded Successfully")

except:
    print("File not found")
```

UNDERSTANDING DATA CONTENT

```python
print("Number of Countries:")

print(df["region_country"].nunique())
```

```python
print("Number of Parameters:")

print(df["parameter"].nunique())

print(df["parameter"].value_counts())
```

```python
print("Year Coverage:")

print(
    df.groupby("category")["year"]
      .agg(["min", "max"])
)
```

```python
# Removing duplicates

df = df.drop_duplicates()
```

```python
# Standardizing column names

df.columns = (

    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```
