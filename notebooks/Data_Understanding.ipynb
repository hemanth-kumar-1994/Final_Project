```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

Load the IEA Global EV Outlook 2025 dataset and inspect its structure before preprocessing.

```python
file_path = "../data/raw/EV Data Explorer 2025.xlsx"

Excel_sheets = pd.ExcelFile(file_path)

print(Excel_sheets.sheet_names)
```

READING MAIN SHEET FROM WORKBOOK

```python
df = pd.read_excel(file_path, sheet_name="GEVO_EV_2025")
```

DATASET STRUCTURE

The dataset contains 16436 observations and 9 variables, indicating comprehensive country-level EV data.

```python
print(df.shape)

print(df.info())
```

```python
df.head()
```

DROPPING DUPLICATE RECORDS

```python
# Check duplicate records
print("Duplicate rows before removal:", df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

# Verify removal
print("Duplicate rows after removal:", df.duplicated().sum())

# Check updated dataset shape
print("Updated dataset shape:", df.shape)
```

EXPLORING SOME USEFUL COLUMNS

```python
columns  = ['region_country', 'category', 'parameter', 'powertrain']
```

```python
print(df[i].value_counts() for i in columns )
```

```python
print("Number of unique countries in the file : ", df["region_country"].nunique())
```

```python
print("Year Coverage:")

df.groupby("category")["year"].agg(["min", "max"])
```

RESTRUCTURING COLUMN NAMES

```python
df.columns = (

    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

SAVING CLEAN DATA

```python
df.to_csv(
    "../data/processed/ev_clean_dataset_2025.csv",
    index=False,
)

print("Clean dataset saved successfully.")
```
