```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

LOADING WORKBOOK


```python
file_path = "../data/raw/EV Data Explorer 2025.xlsx"

Excel_sheets = pd.ExcelFile(file_path)

print(Excel_sheets.sheet_names)
```

READING MAIN SHEET FROM WORKBOOK

```python
data_frame = pd.read_excel(file_path, sheet_name="GEVO_EV_2025")
```

DATA STRUCTURE AND USEFUL INFORMATION

```python
print(data_frame.shape)

print(data_frame.info())
```

```python
data_frame.head()
```

CHECKING DUPLICATE RECORDS

```python
duplicates = data_frame.duplicated().sum()

print(f'Duplicate rows count : {duplicates}')
```

EXPLORING SOME USEFUL COLUMNS

```python
columns  = ['region_country', 'category', 'parameter', 'powertrain']
```

```python
print(data_frame[i].value_counts() for i in columns )
```

```python
data_frame["region_country"].nunique()
```

```python
print("Year Coverage:")

data_frame.groupby("category")["year"].agg(["min", "max"])
```

RESTRUCTURING COLUMN NAMES

```python
data_frame.columns = (

    data_frame.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

data_frame.columns
```

SAVING CLEAN DATA

```python
data_frame.to_csv(
    "../data/processed/ev_clean_dataset_2025.csv",
    index=False
)

print("Clean dataset saved successfully.")
```
