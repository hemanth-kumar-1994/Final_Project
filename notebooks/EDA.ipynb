DELETING DATA WHICH IS NOT RELATED TO COUNTRIES AND CREATING A DATAFRAME

```python
not_countries = [
    "World",
    "Africa",
    "Asia Pacific",
    "Central and South America",
    "EU27",
    "Europe",
    "Middle East and Caspian",
    "North America",
    "Rest of the world"
]
```

```python
country_df = df[
    ~df["region_country"].isin(not_countries)
].copy()
```

```python
ev_sales_df = country_df[
    (country_df["category"] == "Historical") &
    (country_df["parameter"] == "EV sales")
].copy()

print(ev_sales_df.shape)
print(ev_sales_df["mode"].value_counts())
print(ev_sales_df["powertrain"].value_counts())
```

```python
model_df = country_df[
    (country_df["category"] == "Historical") &
    (country_df["parameter"] == "EV sales")
].copy()
```

```python
print(model_df["region_country"].nunique())

print(sorted(model_df["year"].unique()))

print(model_df.duplicated().sum())

model_df.head()
```

```python
country_year_df = (
    model_df
    .groupby(
        ["region_country", "year"],
        as_index=False
    )["value"]
    .sum()
)

print(country_year_df.shape)

country_year_df.head()
```

```python
country_year_df.describe()
```

```python
country_year_df.groupby("region_country").size().describe()
```

```python
country_counts = (
    country_year_df
    .groupby("region_country")
    .size()
    .sort_values()
)

print(country_counts)
```

REMOVING DATA WHICH IS HAVING DATA BELOW 10 YEARS

THERE IS NO UNIVERSAL RULE BUT 10 YEARS IS A COMMON THRESHOLD BECAUSE

ADOPTION PROCESS UNFOLD SLOWLY

IF ONLY FEW COUNTRIES HAVE LESS DATA , THEY BEHAVE LIKE OUTLIERS AND CAN DISTORT MODEL TRAINING



```python
country_counts = country_year_df.groupby("region_country").size()

valid_countries = country_counts[country_counts >= 10].index

model_df = country_year_df[country_year_df["region_country"].isin(valid_countries)].copy()
```

```python
print(model_df.shape)

print(model_df["region_country"].nunique())

print(model_df["year"].min(), model_df["year"].max())
```

```python
model_df.to_csv(
    "../data/processed/eda_ev_dataset.csv",
    index=False
)

print("Clean dataset saved successfully.")
```
